```python
#Task 1
import numpy as np
import matplotlib.pyplot as plt
xPoints=np.array([1,2,3,4,5])
yPoints=np.array([10,20,30,40,50])
xTempPoints=np.array([-1,-2,-3,-4,-5])
yTempPoints=np.array([10,20,30,40,50])
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(xPoints,yPoints,linestyle='dotted')
plt.plot(xTempPoints,yTempPoints)
plt.show()
```

```python
#Task 2
import matplotlib.pyplot as plt
labels=['Mortgage','Repairs','Food','Liabilities']
percent=[51.72,10.34,17.24,20.69]
fig,a=plt.subplots(figsize=(10,7))
explode = (0.1, 0.1, 0.1,0.1)
a.pie(percent,labels=labels,startangle=90,autopct='%.2f%%',shadow=True,explode=explode)
plt.title("Household Expenses")
plt.show()
```
